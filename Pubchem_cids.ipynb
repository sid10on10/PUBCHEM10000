```python
import pubchempy as pcp

with open('drugs.txt', 'r') as file:
    for drugs in file:
        print(drugs)
        results = pcp.get_cids(drugs, 'name')
        #print(results)
        with open('drugs_data.txt', 'a') as drugfile:
            drugfile.write(drugs + "\n")
            if len(results)== '0':
                drugfile.write("None\n")
            else:
                for i in range(len(results)):
                    drugfile.write(str(results[i]) + "\n")
```

This is used at 1st Phase to get cids from Pubchem using Pubchempy library
