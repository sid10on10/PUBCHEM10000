```python
import pubchempy as pcp

with open('drugs.txt', 'r') as file:
    for drugs in file:
        results = pcp.get_cids(drugs, 'name')
        with open('drugs_pubchem_data.txt', 'a') as drugfile:
            drugfile.write(drugs + "\n")
            print(drugs)
            if len(results) == '0':
                drugfile.write("CID:" + "None\n")
            else:
                for i in range(len(results)):
                    drugfile.write("CID:" + str(results[i]) + "\n")
                    c = pcp.Compound.from_cid(str(results[i]))
                    drugfile.write(str(c.iupac_name)+ "\n")
                    print(c.iupac_name)
                    drugfile.write(str(c.molecular_formula)+ "\n")
                    drugfile.write(str(c.isomeric_smiles)+ "\n")
                    drugfile.write(str(c.molecular_weight)+ "\n")
                    drugfile.write(str(c.isomeric_smiles)+ "\n")
                    drugfile.write(str(c.canonical_smiles)+ "\n")
                    drugfile.write(str(c.inchi)+ "\n")
                    drugfile.write(str(c.xlogp)+ "\n")
                    drugfile.write(str(c.inchikey)+ "\n")
                    drugfile.write(str(c.bonds)+ "\n")
                    drugfile.write(str(c.rotatable_bond_count)+ "\n")
                    drugfile.write(str(c.fingerprint)+ "\n")
                    drugfile.write(str(c.h_bond_acceptor_count)+ "\n")
                    drugfile.write(str(c.h_bond_donor_count)+ "\n")
            
```
